# Columns Details

Serial No.: This column represents a unique serial number or identifier for each record in the dataset.
    
GRE Score: This column denotes the Graduate Record Examination (GRE) score of the applicant, which is a standardized test used for admissions to graduate programs.(290 to 340)
    
TOEFL Score: This column represents the Test of English as a Foreign Language (TOEFL) score of the applicant, which measures English language proficiency.(92 to 120)
    
University Rating: This column indicates the rating or reputation of the university where the applicant is seeking admission.
(1 to 5)

SOP: This column represents the Statement of Purpose (SOP) score, which reflects the quality of the applicant's essay or statement explaining their academic and career goals.(1 to 5)
    
LOR: This column denotes the Letter of Recommendation (LOR) score, which reflects the strength and quality of the recommendation letters provided by referees.(1 to 5)
    
CGPA: This column represents the Cumulative Grade Point Average (CGPA) of the applicant, which is a measure of their academic performance.(6.8 to 9.92)
    
Research: This column is a binary indicator (e.g., 0 or 1) that represents whether the applicant has research experience.(0 or 1)
    
Chance of Admit: This column indicates the likelihood or probability of the applicant being admitted to the university or program they have applied to.(0.34 to 0.97)

## Training and Testing Data

```python
x = dataframe[['TOEFL_Score','CGPA',"SOP","LOR",'University_Rating','GRE_Score','Research']]
y = dataframe['Chance_of_Admit']
```

```python
x.head(2)
```

```python
y.head(2)
```

### Split Data into 80:20

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.20)
print("size of x training dataset ",X_train.shape)
print("size of y training dataset ",y_train.shape)
print("size of x testing dataset ",X_test.shape)
print("size of y testing dataset ",y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_scaler = scaler.fit_transform(X_train)
X_test_scaler = scaler.fit_transform(X_test)

X_test_scaler.shape , X_train_scaler.shape
```

## Training the Model using Linear regression

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression() 
lm.fit(X_train_scaler,y_train)
LinearRegression()
```

```python
print("oceff is : ",lm.coef_)
print("intercept is : ",lm.intercept_)
```

## Predicting Test Data

```python
y_pred = lm.predict(X_test_scaler)
```

## Evaluating the Model


```python
from sklearn import metrics
```

```python
print('MAE :'," ", metrics.mean_absolute_error(y_test,y_pred))
print('MSE :'," ", metrics.mean_squared_error(y_test,y_pred))
print('RMAE :'," ", np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(y_test,y_pred)
score
```

```python
Adj_score = 1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
Adj_score
```

```python
print("r2 square of the model(accuracy of the model) :",score)
print("adjested r2 square :", Adj_score)
```
